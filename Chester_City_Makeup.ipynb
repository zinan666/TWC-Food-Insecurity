```
# Store location

latitude = 39.853886
longitude = -75.358658
data = store[store.City == 'Chester']

# Instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# Loop through the 200 crimes and add each to the incidents feature group
for lat, lng, in zip(data.Y, data.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add incidents to map
chester_map = folium.Map(location=[latitude, longitude], zoom_start=13)
chester_map.add_child(incidents)
latitudes = list(data.Y)
longitudes = list(data.X)
labels = list(data.Store_Name)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(chester_map)    
    
for i in range(len(ches_df['coords'])):
  neighborhood = []
  for j in ches_df['coords'][i]:
    neighborhood.append(list(j[::-1]))
  folium.PolyLine(locations=neighborhood,color="black", weight=2.5, opacity=1).add_to(chester_map)

# add incidents to map
chester_map.add_child(incidents)
```

```
from geojson import Feature, FeatureCollection, Polygon

# extract latitude and longitude information
a= {}
for i in range(len(ches_df)):
  b = []
  for j in ches_df['coords'][i]:
    b.append(list(j))
  a[i] = b

features = []  # geojson information for each region
for i in range(len(a)): 
    circle = a[i]
    polygon = Polygon([circle]) 
    features.append(Feature(
        id=i, #match with region in dataframe
        geometry=polygon 
    ))
feature_collection = FeatureCollection(features) 
```

```
# mean income

chester_map = folium.Map(location=[latitude, longitude], zoom_start=14,control_scale=True)

folium.Choropleth(
    geo_data=feature_collection, 
    name='choropleth',
    data=ches_df, 
    columns=['id','income_per_capita'], 
    key_on='feature.id', 
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,  
    legend_name='Rank' 
).add_to(chester_map)
chester_map
```
